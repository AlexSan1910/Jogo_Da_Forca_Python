```python
import random
from os import system, name

# Função para limpar a tela a cada execução
def limpar_tela():

    # limpar tela windows
    if name == 'nt':
        _ = system('cls')
    # Limpar tela do linux ou mac
    else:
        _ = system('clear')

# Função que desenha a forca na tela
def display_hangman(chances):

    # Lista de estágios da forca
    stages = [  # estágio 6 (final)
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                """,
                # estágio 5
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     /
                   -
                """,
                # estágio 4
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |
                   -
                """,
                # estágio 3
                """
                   --------
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |
                   -
                """,
                # estágio 2
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |
                   -
                """,
                # estágio 1
                """
                   --------
                   |      |
                   |      O
                   |
                   |
                   |
                   -
                """,
                # estágio 0
                """
                   --------
                   |      |
                   |
                   |
                   |
                   |
                   -
                """
    ]
    return stages[chances]

def game():

    limpar_tela()
    palavras = []
    animais = [
          "Cachorro",
          "Gato",
          "Elefante",
          "Leão",
          "Tigre",
          "Girafa",
          "Macaco",
          "Cavalo",
          "Pássaro",
          "Cobra",
          "Peixe",
          "Tubarão",
          "Coelho",
          "Rato",
          "Pato",
          "Águia",
          "Baleia",
          "Zebra",
          "Urso",
          "Raposa",
          "Jacaré",
          "Pinguim",
          "Galinha",
          "Vaca",
          "Golfinho",
          "Formiga",
          "Abelha",
          "Caranguejo",
          "Caracol",
          "Coruja"
      ]

    paises = [
          "Estados Unidos",
          "China",
          "Índia",
          "Rússia",
          "Japão",
          "Alemanha",
          "Reino Unido",
          "França",
          "Brasil",
          "Itália",
          "Canadá",
          "Austrália",
          "Espanha",
          "México",
          "Coreia do Sul",
          "Indonésia",
          "Arábia Saudita",
          "Turquia",
          "África do Sul",
          "Argentina",
          "Nigéria",
          "Paquistão",
          "Bangladesh",
          "Colômbia",
          "Egito"
      ]

    frutas =  [
          "Maçã",
          "Banana",
          "Laranja",
          "Morango",
          "Uva",
          "Pera",
          "Abacaxi",
          "Melancia",
          "Limão",
          "Kiwi",
          "Manga",
          "Pêssego",
          "Cereja",
          "Abacate",
          "Framboesa",
          "Melão",
          "Amora",
          "Coco",
          "Pitaya",
          "Maracujá",
          "Goiaba",
          "Tangerina",
          "Caqui",
          "Ameixa",
          "Figo",
          "Mirtilo",
          "Jabuticaba",
          "Romã",
          "Carambola",
          "Physalis"
      ]

    corpo_humano = [
          "Cabeça",
          "Pescoço",
          "Ombro",
          "Braço",
          "Cotovelo",
          "Antebraço",
          "Pulso",
          "Mão",
          "Dedo",
          "Polegar",
          "Barriga",
          "Cintura",
          "Quadril",
          "Nádegas",
          "Perna",
          "Coxa",
          "Joelho",
          "Panturrilha",
          "Tornozelo",
          "Pé",
          "Calcanhar",
          "Dedo do pé",
          "Pele",
          "Cabelo",
          "Olho",
          "Sobrancelha",
          "Cílio",
          "Orelha",
          "Nariz",
          "Bochecha",
          "Lábio",
          "Língua",
          "Dente",
          "Queixo",
          "Pescoço",
          "Pescoço",
          "Testa"
      ]

    instrumentos = [
          "Violino",
          "Violoncelo",
          "Contrabaixo",
          "Viola",
          "Flauta",
          "Clarinete",
          "Saxofone",
          "Trompete",
          "Trombone",
          "Tuba",
          "Oboé",
          "Fagote",
          "Harpa",
          "Piano",
          "Teclado",
          "Órgão",
          "Guitarra",
          "Baixo elétrico",
          "Violão",
          "Banjo",
          "Ukulele",
          "Bateria",
          "Pandeiro",
          "Congas",
          "Bongô",
          "Timbales",
          "Tamborim",
          "Reco-reco",
          "Berimbau",
          "Acordeão",
          "Sitar",
          "Cítara",
          "Gaita de boca",
          "Xilofone",
          "Marimba",
          "Vibrafone",
          "Tímpanos"
      ]

    # Saudações ao Jogador e escolha de categoria
    print("\nBem-Vindo(a) ao jogo da forca!")

    escolha_user = int(input("""
    Temos algumas categorias, qual mais te interessa ?

    1 - Animais
    2 - Frutas
    3 - Pais
    4 - Partes do corpo humano
    5 - Instrumento Musicais

    """))

    if escolha_user == 1:
      palavras += animais
    elif escolha_user == 2:
      palavras += frutas
    elif escolha_user == 3:
      palavras += paises
    elif escolha_user == 4:
      palavras += corpo_humano
    elif escolha_user == 5:
      palavras += instrumentos
    else:
      print("Você não digitou uma das opções!")


    print("Adivinhe a palavra abaixo: \n")

    # Listas de palavras para o jogo
    # palavras = ["banana", "abacate", "uva", "morango", "laranja"]

    # Escolher de forma randomica uma palavra da lista de palavras
    palavra = random.choice(palavras)

    # Utilizando uma List-Comprehension para mostrar os underline equivalente a quantidade de palavra escolhida
    letras_descobertas = ["_" for letra in palavra]

    # Variavel que irá conter o numero de chances
    chances = 6

    # Variavel para armazenar as letras erradas
    letras_erradas = []

    while chances > 0:

        print(display_hangman(chances))
        print(" ".join(letras_descobertas)) # Mostrando na tela do usuario o list-Comprehension
        print("\nChances restantes: ", chances) # Mostrando a cada loop a quantidade de chances que o usuario possui ainda
        print("Letras erradas:", " ".join(letras_erradas)) # Mostrando a cada loop as letras que o usuario digitou errado

        #Tentativa
        tentativa = input("\nDigite uma letra: ").lower()

        # Condicional
        if tentativa in letras_erradas:
            print("Você já tentou essa letra. Escolha outra!")
            continue

        # Condicional se letra digitado pelo usuario estiver dentro de palavra.
        if tentativa in palavra:
            index = 0

            # para cada letra na variavel palavra, estarei verificando se a letra digitada pelo usuario é igual a letra da palavra
            for letra in palavra:
                # Se a letra digitada pelo usuario "tentativa" for igual a alguma letra da palavra, ele vai pegar essa letra e colocar
                # na mesma posição em que as letras iguais batem.
                if tentativa == letra:
                    letras_descobertas[index] = letra
                index += 1

        # Quando a letra que o usuario digitar não estiver dentro de palavra
        else:
            chances -= 1 # retirando da variavel chance uma "vida", pois ele digitou uma letra errada
            letras_erradas.append(tentativa) # colocando dentro de letras_erradas as letras que o usuario digitou errado


        if "_" not in letras_descobertas:
            print("\n Parabens a palavra era: ", palavra)

    # Quando o loop while for falso, ou seja, acabar as chances do jogador.. então ele irá cair nessa condição
    # Basicamente nessa condição  se houver o urderline ainda em letras_descobertas, mostre uma mensagem ao usuario que ele perdeu
    if "_" in letras_descobertas:
        print("\n Você perdeu o jogo, a palavra era: ", palavra)


# Bloco Main
if __name__ == "__main__":
    game()
```
